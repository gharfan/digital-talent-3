File volumeproduksiperikanan.csv merupakan file yang berisikan data Volume dan nilai produksi perikanan tangkap di laut dan perairan umum menurut jenis ikan dan provinsi dengan deskripsi Kolom sebagai berikut:

# 2 Statistik Data

```python
import pandas as pd
data_produksi_perikanan = pd.read_csv('volumeproduksiperikanan.csv', delimiter =',', encoding = "utf-8")
data_produksi_perikanan
```

```python
data_produksi_jatim = data_produksi_perikanan[data_produksi_perikanan['NamaProvinsi']=='Jawa Timur']
print(data_produksi_jatim)
```

```python
#volume perikanan tangkap terbesar di Jawa Timur
volume_terbesar = data_produksi_jatim.Volume.max()
print("volume perikanan tangkap terbesar di Jawa Timur adalah", volume_terbesar)

#nilai perikanan tangkap terbesar di Jawa Timur
nilai_terbesar = data_produksi_jatim.Nilai.max()
print("nilai perikanan tangkap terbesar di Jawa Timur adalah", nilai_terbesar)

#volume perikanan tangkap terbesar di Indonesia
volume_max = data_produksi_perikanan.Volume.max()
print("volume perikanan tangkap terbesar di Indonesia adalah", volume_max)

#nilai perikanan tangkap terbesar di Indonesia
nilai_max = data_produksi_perikanan.Nilai.max()
print("nilai perikanan tangkap terbesar di Indonesia adalah", nilai_max)

#perbandingan volume perikanan tangkap Jawa Timur dan Indonesia
persentase_volume = volume_terbesar / volume_max *100
print("volume perikanan tangkap Jawa Timur sebesar", persentase_volume, "dari total volume perikanan tangkap di Indonesia")

#perbandingan nilai perikanan tangkap Jawa Timur dan Indonesia
persentase_nilai = nilai_terbesar / nilai_max *100
print("nilai perikanan tangkap Jawa Timur sebesar", persentase_nilai, "dari total nilai perikanan tangkap di Indonesia")
```

# 3 Visualisasi Data

```python
ax = data_produksi_perikanan.groupby('NamaProvinsi').sum()['Volume']
ax = ax.plot(kind='bar', title="Jumlah Volume Perikanan Tangkap berdasarkan Nama Provinsi di Tahun 2000-2012", figsize=(15,5), legend=True, fontsize=12)
ax.set_xlabel("NamaProvinsi", fontsize=12)
ax.set_ylabel("Volume", fontsize=12)
```

```python
ax = data_produksi_perikanan.groupby('NamaProvinsi').sum()['Nilai']
ax = ax.plot(kind='bar', title="Jumlah Nilai Perikanan Tangkap berdasarkan Nama Provinsi Tahun 2000-2012", figsize=(15,5), legend=True, fontsize=12)
ax.set_xlabel("NamaProvinsi", fontsize=12)
ax.set_ylabel("Nilai", fontsize=12)
```

```python
ax = data_produksi_jatim.groupby('Tahun').sum()['Volume']
ax = ax.plot(kind='line', title="Jumlah Volume Perikanan Tangkap di Provinsi Jawa Timur pada Tahun 2000-2012", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Volume", fontsize=12)
```

```python
ax = data_produksi_jatim.groupby('Tahun').sum()['Nilai']
ax = ax.plot(kind='line', title="Jumlah Nilai Perikanan Tangkap di Provinsi Jawa Timur pada Tahun 2000-2012", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai", fontsize=12)
```
